8. Сравните эффективность алгоритмов двоичного поиска, поиска в глубину и поиска в ширину с помощью O-нотации.


```python
def binary_search(arr, target):
    """
    Бинарный поиск в отсортированном массиве.
    Возвращает индекс элемента или -1, если не найден.
    """
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


def search(graph, start, get_next):
    """
    Универсальный алгоритм обхода графа.
    graph: dict[node, list[neighbors]]
    start: стартовая вершина
    visit: функция, вызываемая при посещении вершины
    get_next: стратегия выбора следующего элемента из очереди
    """
    visited = set()
    queue = [start]

    while queue:
        node = get_next(queue)  # pop или pop(0)
        if node not in visited:
            visited.add(node)
            # добавляем соседей
            for neighbor in graph.get(node, []):
                if neighbor not in visited:
                    queue.append(neighbor)


def dfs(graph, start):
    return search(graph, start, lambda q: q.pop()) # LIFO


def bfs(graph, start):
    return search(graph, start, lambda q: q.pop(0)) # FIFO


# Пример
g = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print(dfs(g, 'A'))
print(bfs(g, 'A'))
print(binary_search([1, 3, 5, 7, 9], 7))  # 3
```
